### Las variables
1. NUM = Numero ingresado
2. R = Estado del numero

## Formulas
$R = "NUM >= 0" o "NUM <= 0"$

## Diagrama de flujo en LucidChart
![](https://raw.githubusercontent.com/calebdorado128/Act-12-p3.2-pag-53/refs/heads/main/DF%20if%20else%20Positivo%20Negativo%202.JPG)

```python
# Para cuando sea verdadero
# Calcular estado con if else
# Dato de Entrada: NUM
# Mostrar "Negativo" si es falso
# Si es verdadero mostrar "Positivo"
NUM = int(input("Ingrese un numero: "))
if NUM >= 0:
    R = "Positivo"
else:
    R = "Negativo"
print(R)
print("Fin del programa")
print("Programa realizado por Caleb Dorado 0039")
```
